```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
import warnings
warnings.filterwarnings("ignore")
#импортируем библиотеки и игнорируем предупреждения
```

# Выбрал экспоненциальное распределение - f(x,λ) = 0, x<0 and λ*$e^{-λx}$, x>=0. E[x] = 1/λ, D[x] = 1/$λ^2$

```python
x = np.linspace(1,10,1000) #определяем массив точек для построения плотности распределения
lambd = 0.6 #выбрал λ
expon = sts.expon(lambd) #задаем функцию
sample = expon.rvs(1000) #генерируем случайную выборку 1000 значений
pdf = expon.pdf(x) #вычисляем значение плотности распределения от выборки
```

```python
#строим гистрограмму выборки и теоретическую функцию распределения
plt.hist(sample, bins = 50, range=(1,10), normed = True)
plt.ylabel('$f(x)$/number of samples')
plt.xlabel('$x$')
plt.plot(x, pdf, color ='red', label='theoretical PDF')
plt.legend(loc='best')
```

```python
#посчитаем среднее и дисперсию
E = round(1/lambd, 2) #среднее
D = round(1/lambd**2, 2) #дисперсия

print('E[x] =', E)
print('D[x] =', D)
```

# Оценим распределение выборочного среднего случайной величины при разных объёмах выборок, построим соответствующие графики.

```python
n = 5
sample_new = np.zeros(1000) #создали массив для нашей будущей средней выборки
for i in range(0,n):
    sample_new = expon.rvs(1000)+sample_new #суммируем n случайных выборок
sample_new = sample_new/n #усреднили по n
norm_new = sts.norm(1/lambd, (1/lambd**2/n)**0.5) #задали нормальное распределение с E[x]=1/λ, D[x]= 1/λ^2/n
pdf_new = norm_new.pdf(x) #определили плотность нормального распределения с текущими параметрами E[x] и D[x]
plt.plot(x, pdf_new, label='PDF(n = 5)') #строим PDF
plt.hist(sample_new, bins = 50, range=(0,10), normed=True) #строим гистограмму средней выборки
plt.ylabel('number of samples') #подписываем ось y
plt.xlabel('$x$') #подписываем ось х
plt.legend(loc='best') #определяем положение легенды и выводим её
```
