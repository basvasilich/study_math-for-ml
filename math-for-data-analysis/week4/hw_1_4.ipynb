```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
%matplotlib inline
```

Зададим равномерное распределение

```python
a = 4
b = 10
rv = stats.uniform(a, b - a)
sample = rv.rvs(1000)
```

```python
x = np.linspace(2, 12, 1000)
cdf = rv.cdf(x) # cdf возвращает значение функции распределенияслучайной величины в точке
plt.plot(x, cdf, label='theoretical CDF')

from statsmodels.distributions.empirical_distribution import ECDF
ecdf = ECDF(sample)
plt.step(ecdf.x, ecdf.y, label='ECDF')

plt.ylabel('$F(x)$')
plt.xlabel('$x$')
plt.legend(loc='upper left')
```

 Гистограмма для выборки объема 1000 и теоретическое значение плотности

```python
plt.hist(sample, bins=15, density=True)
plt.plot(x, rv.pdf(x), label='theoretical  PDF')
plt.ylabel('fraction of samples')
plt.xlabel('$x$')
plt.legend(loc='upper left')
```

Оценки распределений выборочных средних случайной величины при разных объёмах выборок.

```python
volume = [5, 10, 100]
M = (a + b) / 2
D = (b - a)**2 / 12 
for n in volume:
    arr = []
    for i in range(10000):
        arr.append(sum(rv.rvs(n))/n)
    print("Гистограмма распределения выборочных средних при объеме {0} и график плотности соответсвующего нормального распределения:"\
          .format(n))
    plt.hist(arr, bins=150, density=True)
    plt.plot(x, stats.norm(M, np.sqrt(D/n)).pdf(x))
    plt.show()
    
```

Вывод: при увеличении n мы можем наблюдать приближение нашего распределения к нормальному, что увеличивает точно аппроксимации
