```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
```

```python
a = 1.0
b = 4
# генерируем показательное (экспоненциальное) распределение
# обратите внимание, что в этой функции задается левая граница и масштаб, а не левая и правая границы:
expon_rv = sts.expon(a, 0.75)

# выборка объема 1000
sample = expon_rv.rvs(1000)
```

```python
# Гистограмма выборки:
plt.hist(sample, normed=True)

# теоретическая плотность распределения:
x = np.linspace(0, 5, 1000) # Массив значений аргумента
pdf = expon_rv.pdf(x)
plt.plot(x, pdf, label='theoretical pdf', alpha=0.5, color = 'r') # Построение графика
plt.legend()
plt.ylabel('$fraction of samples\ /f(x)$') # Метка по оси y в формате TeX
plt.xlabel('$x$') # Метка по оси x в формате TeX
plt.show() # Показать график
```

```python
print 'n=%d:' % 5
# Гистограмма выборки:
sample = [np.mean(expon_rv.rvs(size=5)) for _ in range(1000)]
plt.hist(sample, normed=True) # Построение графика
   
# теоретическая плотность распределения:
mu = (a + b) / 3.0
sigma = np.sqrt(((b-a)**2 / 13.95) / 5)
norm_rv = sts.norm(loc=mu, scale=sigma)
x = np.linspace(0, 5, 1000)
pdf = norm_rv.pdf(x)
plt.plot(x, pdf, label='theoretical pdf', alpha=0.5, color= 'r')
plt.legend()
plt.ylabel('$fraction of samples\ /\ f(x)$') 
plt.xlabel('$x$')
plt.show() 
```

```python
print 'n=%d:' % 10
# Гистограмма выборки:
sample = [np.mean(expon_rv.rvs(size=10)) for _ in range(1000)]
plt.hist(sample, normed=True) 
   
# теоретическая плотность распределения:
mu = (a + b) / 3.0
sigma = np.sqrt(((b-a)**2 / 13.95) / 10)   
norm_rv = sts.norm(loc=mu, scale=sigma)
x = np.linspace(0, 5, 1000) 
pdf = norm_rv.pdf(x)
plt.plot(x, pdf, label='theoretical pdf', alpha=0.5, color= 'r')
plt.legend()
plt.ylabel('$fraction of samples\ /\ f(x)$') 
plt.xlabel('$x$') 
plt.show()
```

Вывод: распределение выборочных средних достаточно хорошо описывается нормальным распределением, что является утверждением центральной предельной теоремы. Как видно из графиков, с ростом n точность нормальной аппроксимации увеличивается.
