```python
# Импортируем нужные библотеки
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

```python
# Выберите ваше любимое непрерывное распределение
# Распределение Максвела

# Сгенерируйте из него выборку объёма 1000
sample = sts.maxwell.rvs(size=1000)
```

```python
# постройте гистограмму выборки и нарисуйте поверх неё теоретическую плотность распределения вашей случайной величины
plt.hist(sample, bins=15, density=True)
plt.ylabel('Fraction of samples')
plt.xlabel('$x$')
```

```python
# при трёх и более значениях n (например, 5, 10, 50) сгенерируйте 1000 выборок объёма n
n_sizes = np.array([5,10,50])
num_of_samples = 1000

samples = {}

for n in n_sizes:
    samples[n] = []
    for j in range(num_of_samples):
        samples[n].append(sts.maxwell.rvs(size=n))
        
samples
```
